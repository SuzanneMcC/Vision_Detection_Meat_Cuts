```
%reload_ext autoreload
%autoreload 2
%matplotlib inline

from fastai import *
from fastai.vision import *
from fastai.metrics import error_rate, FBeta
from fastai.callbacks import *
#import pickle
```

```
data1 = ImageDataBunch.from_folder(path='/content/drive/MyDrive/FastAI/Meat New Dataset Colour TTV', 
                                  train = 'train', 
                                  valid = 'val',
                                  size =64,
                                  bs=64)
```

```
data2 = ImageDataBunch.from_folder(path='/content/drive/MyDrive/FastAI/Meat New Dataset Colour TTV', 
                                  train = 'train', 
                                  valid = 'test',
                                  size =64,
                                  bs=64)
```

```
data1.classes
```

```
data1.show_batch(rows = 4, figsize=(10,10))
```

```
model1= cnn_learner(data1,models.resnet34,metrics=[accuracy,
               Precision(average='micro'),
               Recall(average='micro'),
               FBeta(average='micro')], callback_fns= ShowGraph)
```

```
model1.fit_one_cycle(50, callbacks =[callbacks.SaveModelCallback(model1, every='epoch', monitor = 'valid_loss')])
```

```
from fastai.core import *

model2= cnn_learner(data2,models.resnet34,metrics=[accuracy,
               Precision(average='micro'),
               Recall(average='micro'),
               FBeta(average='micro')], callback_fns= ShowGraph)
```

```
model2.load('/content/drive/MyDrive/FastAI/Meat New Dataset Colour TTV/models/bestmodel_29')
```

```
model2.validate(data2.valid_dl)
```
