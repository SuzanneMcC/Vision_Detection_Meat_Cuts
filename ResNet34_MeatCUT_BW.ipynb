```
%reload_ext autoreload
%autoreload 2
%matplotlib inline

from fastai import *
from fastai.vision import *
from fastai.metrics import error_rate, FBeta
from fastai.callbacks import *
```

```
data = ImageDataBunch.from_folder(path='/content/drive/MyDrive/FastAI/Meat New Dataset BW TTV', 
                                  train = 'train', 
                                  valid = 'val',
                                  size =64,
                                  bs=64)
#dls = ImageDataLoaders.from_folder(path='/content/drive/My Drive/Meat_Images/TrainTest',
                                   #valid_pct=0.2)
```

```
data_new = ImageDataBunch.from_folder(path='/content/drive/MyDrive/FastAI/Meat New Dataset BW TTV', 
                                  train = 'train', 
                                  valid = 'test',
                                  size =64,
                                  bs=64)
```

```
data.classes
```

```
data.show_batch(rows = 4, figsize=(10,10))
```

```
model= cnn_learner(data,models.resnet34,metrics=[accuracy,
               Precision(average='micro'),
               Recall(average='micro'),
               FBeta(average='micro')], callback_fns= ShowGraph)
```

```
#model.unfreeze() 
#model.fit_one_cycle(3, max_lr=slice(1e-03, 1e-02))
model.fit_one_cycle(50, callbacks =[callbacks.SaveModelCallback(model, every='epoch', monitor = 'valid_loss')])
```

```
model_new= cnn_learner(data_new,models.resnet34,metrics=[accuracy,
               Precision(average='micro'),
               Recall(average='micro'),
               FBeta(average='micro')], callback_fns= ShowGraph)
```

```
model_new.load('/content/drive/MyDrive/FastAI/Meat New Dataset BW TTV/models/bestmodel_48')
```

```
model_new.validate(data_new.valid_dl)
```
